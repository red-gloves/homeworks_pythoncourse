# Тема №1

Создайте класс Recipe с полями название, ингредиенты и время приготовления. Реализуйте методы update_cooking_time(new_time), который обновляет время приготовления, и display_info(), который выводит информацию о рецепте.

```python
class Recipe:
    def __init__(self, name, ingredients, cooking_time):
        self.name = name
        self.ingredients = ingredients
        self.cooking_time = cooking_time

    def update_cooking_time(self, new_time):
        if new_time >= 0:
            self.cooking_time = new_time
        else:
          print("Время приготовления не может быть отрицательным или равным нулю.")

    def display_info(self):
        print(f"Название: {self.name}, Ингредиенты: {self.ingredients}, Время приготовления: {self.cooking_time}")

fried_chicken = Recipe("Жареная курица", "Филе курицы, масло, соль", 90)
fried_chicken.display_info()
fried_chicken.update_cooking_time(120)
fried_chicken.display_info()
```

# Тема №2

Определите класс Gadget, который наследуется от класса Product. Класс Gadget должен добавлять методы для обновления цены и отображения информации о гаджете.

```python
class Product:
    def __init__(self, name, price):
      self.name = name
      self.price = price
    def display_info(self):
      print(f"Name: {self.name}, Price: {self.price}")

class Gadget(Product):
    def __init__(self, name, price, model):
      super().__init__(name, price)
      self.model = model

    def update_price(self, new_price):
      self.price = new_price

    def display_gadget_info(self):
      self.display_info()
      print(f"Model: {self.model}")

smartphone = Gadget("Samsung Galaxy", "100$", "S10+")
smartphone.display_gadget_info()
smartphone.update_price("80$")
smartphone.display_gadget_info()
```

# Тема №4.1

Парк и аттракцион.  Создайте класс `Парк`, который включает в себя аттракционы. У класса `Парк` должен быть метод для добавления аттракционов. Класс `Аттракцион` должен иметь поле `название` и метод для запуска аттракциона. Аттракционы не могут существовать вне парка.

```python
class Park:
  def __init__(self, name, address):
    self.name = name
    self.address = address
    self.attractions = []

  def add_attraction(self, attraction_name, capacity):
    new_attraction = self.Attraction(attraction_name, capacity)
    self.attractions.append(new_attraction)

  def print_info(self):
    print(f"Парк: {self.name}")
    print(f"Адрес: {self.address}")
    print("Список аттракционов:")
    for attraction in self.attractions:
        print(f"- {attraction.attraction_name}, кол-во мест - {attraction.capacity}")

  class Attraction:
    def __init__(self, attraction_name, capacity):
      self.attraction_name = attraction_name
      self.capacity = capacity

    def start(self):
      print(f"Аттракцион '{self.attraction_name}' запущен!")

park = Park("Disneyland Paris", "Marne-la-Vallée")

park.add_attraction("Ferris wheel", "2160")
park.add_attraction("Roller coaster", "40")

park.print_info()

park.attractions[0].start()
```

Тур и экскурсовод. Создайте класс `Тур`, который может проводиться разными экскурсоводами. У класса `Экскурсовод` есть поля `имя` и `языки`. Экскурсоводы могут существовать независимо от тура.

```python
class Guide:
  def __init__(self, name, langs):
    self.name = name
    self.langs = langs

  def get_name(self):
    return self.name

  def get_langs(self):
    return self.langs

  def display_info(self):
    print(f"Имя: {self.name}, Языки: {self.langs}")

class Tour:
  def __init__(self):
    self.guides = []

  def add_guide(self, guide):
    self.guides.append(guide)

  def display_guides(self):
    print("Список экскурсоводов тура: ")
    for guide in self.guides:
      guide.display_info()


guide1 = Guide("Анна", "русский, английский")
guide2 = Guide("Мария", "русский, немецкий, португальский")
guide3 = Guide("Ольга", "русский, французский, китайский, польский")

tour1 = Tour()

tour1.add_guide(guide1)
tour1.add_guide(guide2)
tour1.add_guide(guide3)

tour1.display_guides()
```

# Тема №6

Создать интерфейс IRecipe, который содержит методы updatePreparationTime(int time) и getRecipeDetails(). Создать класс Recipe, реализующий этот интерфейс. Класс должен содержать поля title, ingredients и preparationTime.

```python
from abc import ABC, abstractmethod

class IRecipe(ABC):
  @abstractmethod
  def updatePreparationTime(self, time):
    pass

  @abstractmethod
  def getRecipeDetails(self):
    pass

class Recipe(IRecipe):
  def __init__(self, title, ingredients, preparationTime):
    self.title = title
    self.ingredients = ingredients
    self.preparationTime = preparationTime

  def updatePreparationTime(self, new_time):
    self.preparationTime = new_time

  def getRecipeDetails(self):
    print(f"Title: {self.title}, Ingredients: {self.ingredients}, Preparation time: {self.preparationTime}")


recipe1 = Recipe("Fried chicken", "Chicken, salt, oil", "90 minutes")
recipe1.getRecipeDetails()
recipe1.updatePreparationTime("100 minutes")
recipe1.getRecipeDetails()
```
