1) Найти количество и сумму нечетных элементов.

```python
N = int(input())
nums = []
counter = 0
for i in range(N):
  a = int(input())
  nums.append(a)

print("\n")

for i in range(N):
  if nums[-i - 1] % 2 == 0:
    print(nums[-i-1])
    counter += 1
print(f'Количество чётных чисел в массиве: {counter}')
```

3)Дан массив размера N и целое число K (1<=K < N). Осуществить сдвиг элементов массива вправо на K позиций (при этом A1 перейдет в AK+1, A2 — в AK+2, …, AN–K — в AN, а исходное значение K последних элементов будет потеряно). Первые K элементов полученного массива положить равными 0.

```python
N = int(input())
K = int(input())
nums = []
for i in range(N):
  a = int(input())
  nums.append(a)
result = [0] * K + nums[:N-K]

print(result)
```

4) Дан целочисленный двумерный массив, размерности n х m. Выяснить, какое число встречается в какой строке раньше — положительное или отрицательное.

```python
n, m = map(int, input().split())
nums = []
for i in range(n):
  nums.append(list(map(int, input().split())))
for i in range(n):
  for j in range(m):
      if nums[i][j] > 0:
          print(f"Строка {i + 1}: Положительное встретилось раньше", )
          break
      elif nums[i][j] < 0:
          print(f"Строка {i + 1}: Отрицательное встретилось раньше")
          break
```

# Dictionary

1) Подсчет частоты слов: Напишите функцию, которая принимает строку и возвращает словарь, где ключи — это слова, а значения — их частота в строке.

2) Инвертирование словаря: Напишите функцию, которая принимает словарь и возвращает новый словарь, где ключи и значения поменяны местами. Если значения не уникальны, объедините ключи в список.

3) Слияние словарей: Напишите функцию, которая принимает два словаря и возвращает новый словарь, содержащий все ключи из обоих словарей. Если ключи совпадают, значения должны быть суммированы.

4) Фильтрация по значению: Напишите функцию, которая принимает словарь и пороговое значение, и возвращает новый словарь, содержащий только те пары ключ-значение, где значение больше порогового.

5) Максимальное значение: Напишите функцию, которая находит ключ с максимальным значением в словаре.

6) Минимальное значение: Напишите функцию, которая находит ключ с минимальным значением в словаре.

7) Удаление ключей: Напишите функцию, которая принимает словарь и список ключей, и удаляет из словаря все указанные ключи.

8) Группировка по длине ключей: Напишите функцию, которая группирует ключи словаря по их длине и возвращает словарь, где ключи — это длины, а значения — списки ключей соответствующей длины.

9) Поиск ключей по значению: Напишите функцию, которая принимает словарь и значение, и возвращает список всех ключей, связанных с этим значением.

10) Объединение значений: Напишите функцию, которая принимает два словаря и объединяет их значения в списки для каждого ключа.

```python
def word_frequency(text):
    d = {}
    for word in text.lower().split():
        word = ''.join(c for c in word if c.isalpha())
        if word: d[word] = d.get(word, 0) + 1
    return d

def invert_dict(d):
    inv = {}
    for k, v in d.items():
        inv[v] = inv.get(v, []) + [k]
    return inv

def merge_dicts(d1, d2):
    return {k: d1.get(k,0) + d2.get(k,0) for k in set(d1)|set(d2)}

def filter_dict(d, t):
    return {k:v for k,v in d.items() if v>t}

def max_key(d):
    return max(d, key=d.get)

def min_key(d):
    return min(d, key=d.get)

def remove_keys(d, keys):
    return {k:v for k,v in d.items() if k not in keys}

def group_keys_by_length(d):
    g = {}
    for k in d:
        g.setdefault(len(str(k)), []).append(k)
    return g

def find_keys_by_value(d, val):
    return [k for k,v in d.items() if v==val]

def merge_values(d1, d2):
    return {k: [d1.get(k), d2.get(k)] for k in set(d1)|set(d2)}

text = "hello world hello python world"
print("1. Частота слов:", word_frequency(text))

d = {'a':1, 'b':2, 'c':2}
print("2. Инвертированный словарь:", invert_dict(d))

print("3. Объединение:", merge_dicts({'a':1,'b':2}, {'b':3,'c':4}))
print("4. Фильтрация:", filter_dict({'a':1,'b':2,'c':3}, 1))
print("5. Ключ макс:", max_key({'a':1,'b':3,'c':2}))
print("6. Ключ мин:", min_key({'a':1,'b':3,'c':2}))
print("7. Удаление:", remove_keys({'a':1,'b':2,'c':3}, ['a','c']))
print("8. Группировка:", group_keys_by_length({'apple':1,'banana':2,'kiwi':3}))
print("9. Поиск по значению:", find_keys_by_value({'a':1,'b':2,'c':1}, 1))
print("10. Объединение значений:", merge_values({'a':1,'b':2}, {'b':3,'c':4}))
```

# Множества

Создание множества из списка: Напишите функцию, которая принимает список и возвращает множество, содержащее уникальные элементы списка.

Удаление дубликатов: Напишите функцию, которая удаляет дубликаты из списка с помощью множества и возвращает новый список.

Пересечение множеств: Напишите функцию, которая принимает два множества и возвращает их пересечение.

Объединение множеств: Напишите функцию, которая принимает два множества и возвращает их объединение.

Разность множеств: Напишите функцию, которая принимает два множества и возвращает разность первого множества и второго.

Симметрическая разность: Напишите функцию, которая принимает два множества и возвращает их симметрическую разность.

Проверка подмножества: Напишите функцию, которая проверяет, является ли одно множество подмножеством другого.

Проверка надмножества: Напишите функцию, которая проверяет, является ли одно множество надмножеством другого.

Добавление элементов: Напишите функцию, которая добавляет все элементы из списка в множество.

Удаление элементов: Напишите функцию, которая удаляет из множества все элементы, присутствующие в списке.

```python
def list_to_set(lst):
    return set(lst)

def remove_duplicates(lst):
    return list(set(lst))

def set_intersection(set1, set2):
    return set1 & set2

def set_union(set1, set2):
    return set1 | set2

def set_difference(set1, set2):
    return set1 - set2

def symmetric_difference(set1, set2):
    return set1 ^ set2

def is_subset(set1, set2):
    return set1 <= set2

def is_superset(set1, set2):
    return set1 >= set2

def add_to_set(s, lst):
    s.update(lst)
    return s

def remove_from_set(s, lst):
    s.difference_update(lst)
    return s

my_list = [1, 2, 2, 3, 4, 4, 5]
print("1. Множество из списка:", list_to_set(my_list))
print("2. Список без дубликатов:", remove_duplicates(my_list))

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print("3. Пересечение:", set_intersection(a, b))
print("4. Объединение:", set_union(a, b))
print("5. Разность a - b:", set_difference(a, b))
print("6. Симметрическая разность:", symmetric_difference(a, b))
print("7. Является ли {1,2} подмножеством:", is_subset({1,2}, a))
print("8. Является ли a надмножеством {1,2}:", is_superset(a, {1,2}))

s = {1, 2, 3}
print("9. Добавление элементов:", add_to_set(s, [3, 4, 5]))
print("10. Удаление элементов:", remove_from_set(s, [2, 5]))
```

# Tuple

Создание кортежа из списка: Напишите функцию, которая принимает список и возвращает кортеж, содержащий те же элементы.

Объединение кортежей: Напишите функцию, которая принимает два кортежа и возвращает их объединение.

Проверка на наличие элемента: Напишите функцию, которая проверяет, содержится ли заданный элемент в кортеже.

Индекс первого вхождения: Напишите функцию, которая находит индекс первого вхождения заданного элемента в кортеже.

Количество вхождений элемента: Напишите функцию, которая подсчитывает, сколько раз заданный элемент встречается в кортеже.

Максимальный элемент: Напишите функцию, которая находит максимальный элемент в кортеже.

Минимальный элемент: Напишите функцию, которая находит минимальный элемент в кортеже.

Сортировка кортежа: Напишите функцию, которая возвращает отсортированный список на основе элементов кортежа.

Разбиение кортежа: Напишите функцию, которая разбивает кортеж на две части: первую половину и вторую половину.

Обратный порядок: Напишите функцию, которая возвращает кортеж с элементами в обратном порядке.

```python
def list_to_tuple(lst):
    return tuple(lst)

def combine_tuples(t1, t2):
    return t1 + t2

def is_in_tuple(t, element):
    return element in t

def first_index(t, element):
    return t.index(element)

def count_in_tuple(t, element):
    return t.count(element)

def max_in_tuple(t):
    return max(t)

def min_in_tuple(t):
    return min(t)

def sorted_tuple(t):
    return sorted(t)

def split_tuple(t):
    half = len(t) // 2
    return (t[:half], t[half:])

def reverse_tuple(t):
    return t[::-1]

my_list = [1, 2, 3, 4, 5]
print("1. Кортеж из списка:", list_to_tuple(my_list))

t1 = (1, 2, 3)
t2 = (4, 5, 6)
print("2. Объединение кортежей:", combine_tuples(t1, t2))
print("3. Проверка наличия элемента 3:", is_in_tuple(t1, 3))
print("4. Индекс элемента 2:", first_index(t1, 2))
print("5. Количество элементов 3:", count_in_tuple((1, 2, 3, 3, 2), 3))
print("6. Максимальный элемент:", max_in_tuple(t1))
print("7. Минимальный элемент:", min_in_tuple(t1))
print("8. Отсортированный список:", sorted_tuple((3, 1, 4, 2)))
print("9. Разбиение кортежа:", split_tuple((1, 2, 3, 4, 5, 6)))
print("10. Обратный порядок:", reverse_tuple((1, 2, 3, 4)))
```
