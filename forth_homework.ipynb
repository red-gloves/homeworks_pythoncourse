3) Двумерный массив М*N содержит некоторые буквы русского алфавита, расположенные в произвольном порядке. Написать программу, проверяющую можно ли из этих букв составить данное слово S.


```python
M = int(input())
N = int(input())
alpha = []
for i in range(M):
  alpha.append(input().split())

S = input()

grid_letters = {}
for row in alpha:
    for letter in row:
        if letter in grid_letters:
            grid_letters[letter] += 1
        else:
            grid_letters[letter] = 1

word_letters = {}
for letter in S:
    if letter in word_letters:
        word_letters[letter] += 1
    else:
        word_letters[letter] = 1
flag = True

for letter, count in word_letters.items():
    if letter not in grid_letters or grid_letters[letter] < count:
        flag = False
if flag:
   print('ДА')
else:
  print('НЕТ')
```

# Текстовые файлы: группа Text

1) Записать в файл f последовательного доступа N натуральных чисел. Получить в другом файле последовательного доступа все компоненты файла f, кроме тех, которые кратны K. Вывести полученный файл на печать.

```python
N = int(input("Введите количество чисел (N): "))
K = int(input("Введите число K (исключаем кратные): "))

f = open('f.txt', 'w')
for i in range(1, N + 1):
    f.write(f"{i}\n")
f.close()

f = open('f.txt', 'r')
filtered = open('filtered.txt', 'w')

for line in f:
    num = int(line.strip())
    if num % K != 0:
        filtered.write(f"{num}\n")

f.close()
filtered.close()

print("Содержимое файла filtered.txt:")
filtered = open('filtered.txt', 'r')
for line in filtered:
    print(line.strip())
filtered.close()
```

2) Дан файл, содержащий различные даты. Каждая дата — это число месяц и год. Найти самую позднюю дату.

```python
import random

with open('dates.txt', 'w') as f:
    for _ in range(10):
        day = random.randint(1, 28)
        month = random.randint(1, 12)
        year = random.randint(2000, 2023)
        f.write(f"{day} {month} {year}\n")


latest_date = None

with open('dates.txt', 'r') as file:
    for line in file:
        day, month, year = map(int, line.strip().split())
        current_date = (year, month, day)

        if latest_date is None or current_date > latest_date:
            latest_date = current_date

print("Содержимое файла dates.txt:")
with open('dates.txt', 'r') as file:
    print(file.read())

if latest_date:
    year, month, day = latest_date
    print(f"\nСамая поздняя дата: {day:02d}.{month:02d}.{year}")
else:
    print("Файл не содержит дат")
```

3) Дан файл, содержащий текст на русском языке. Определить, сколько раз встречается в нем самое короткое слово.

```python
import random
from collections import Counter

russian_words = ['привет', 'мир', 'это', 'тест', 'слово', 'код', 'программа',
                'файл', 'текст', 'данные', 'я', 'ты', 'он', 'она', 'мы', 'вы',
                'они', 'дом', 'кот', 'собака', 'год', 'день', 'ночь', 'вода',
                'огонь', 'земля', 'воздух', 'дверь', 'окно', 'стол', 'стул']

with open('text.txt', 'w', encoding='utf-8') as f:
    for _ in range(100):
        word = random.choice(russian_words)
        f.write(word + ' ' + random.choice(['.', ',', '!', '?', ' ']))

with open('text.txt', 'r', encoding='utf-8') as file:
    text = file.read()

words = []
current_word = []
for char in text.lower():
    if char.isalpha():
        current_word.append(char)
    else:
        if current_word:
            words.append(''.join(current_word))
            current_word = []
if current_word:
    words.append(''.join(current_word))

if not words:
    print("Файл не содержит слов")
    exit()

min_length = min(len(word) for word in words)
shortest_words = [word for word in words if len(word) == min_length]
count = len(shortest_words)

print(f"Самое короткое слово имеет длину {min_length} букв(ы)")
print(f"Таких слов найдено {count}:")
word_counts = Counter(shortest_words)
for word, cnt in word_counts.items():
    print(f"'{word}' - {cnt} раз(а)")
```
